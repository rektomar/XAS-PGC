```python
import matplotlib.pyplot as plt
from utils.datasets import load_dataset
```

```python
loaders = load_dataset('qm9xas', 256, [0.8, 0.1, 0.1], standardize='normal')
transform = loaders['transform']
data = loaders['loader_trn'].dataset.data
print(data.shape)
print(data.mean(0), data.std(0))

t_data = transform.inverse(data)
print(t_data.shape)
print(t_data.mean(0), t_data.std(0))
```
