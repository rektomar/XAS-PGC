```python
import torch
import matplotlib.pyplot as plt

from utils.spec_datasets import load_dataset

energies = torch.linspace(270, 300, 100)
loaders = load_dataset('qm9xas', 1024, [0.8, 0.1, 0.1])
```

```python
# normalized spectra
def plot_multiple_spectra(specs, alpha=0.1):
    for spec in specs:
        plt.plot(energies, spec, alpha=alpha, color='blue')
    
    plt.xlabel('Energy')
    plt.show()

batch = next(iter(loaders['loader_trn']))

specs = batch['spec'][:1000]
plot_multiple_spectra(specs)
```

```python
mu, sigma = loaders['transform']

plt.plot(energies, mu, color='blue', label='Mean Spectrum')
plt.fill_between(energies, mu - sigma, mu + sigma, color='blue', label='1 std', alpha=0.3)
plt.xlabel('Energy')
plt.legend()
plt.show()
```
